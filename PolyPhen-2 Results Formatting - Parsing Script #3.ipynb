# Creating a script that accepts the Polyphen-2 submission results as a txt file and coverts it to a CSV file. 

### This enables data manipulation of the results and the ability to add them into the master spreadsheet.

```python
#importing Pandas package
import pandas as pd
```

```python
#creating an empty list for rows
#Each parsed line from the text file will be stored here as a list of values
rows = []

#opening the Polyphen results txt file and read it line by line with a loop
#For All of Us Workbench, you need to upload the txt file into the workspace bucket and then open it to run
#For local file, just set the path to the desired txt file
with open("Poly-phen BRCA1 Missing Variants.txt") as f:
    for line in f:
        if line.startswith("#"):  #If a line starts with "#", the script skips that line and moves to the next one. (This avoids processing the header row)

            continue
        
        parts = line.split() #Splits the line into a list of values using whitespace as the delimiter

        # fix prediction column because it has two words for some rows (ex. possibly damaging), and we need to combine them back into one value. The last step split all values separately
        if parts[9] in ["possibly", "probably"]: 
            parts[9] = parts[9] + " " + parts[10] #puts values possibly damaging and probably damaging back together
            del parts[10] #Deletes the now duplicated "damaging" value, and prevents the columns from being misaligned

        rows.append(parts) #Adds the cleaned row (list of values) to the rows list, then iterates loop, after processing the file, rows becomes a list of all parsed rows

#setting the columns names in the spreadsheet that correspond to the dataset fields
cols = [
"o_acc","o_pos","o_aa1","o_aa2","rsid",
"acc","pos","aa1","aa2",
"prediction","pph2_prob","pph2_FPR","pph2_TPR"
]

#Converts the rows list into a pandas DataFrame and assigns the defined column names
df = pd.DataFrame(rows, columns=cols)
```
