# BRCA 2 Concordance Script

### This script evaluates concordance between PolyPhen-2 and REVEL predictions and also compares each model’s predictions to the variant’s clinical significance reported in ClinVar

There are also concordance check between student predictions. We had each student predict with PolyPhen-2 to add redunancy into our validation and make sure everyone was getting the same results.

```python
#reading in the completed file
#this file is the shared V8 BRCA2 file that everyone worked on
#you may need to save the excel spreadsheet into a csv to use for AOU.
#If mulitple tabs, just copy the desired tab into a spreadsheet and save as CSV. Only values are kept, not formating

#Reading in our data, if in AOU just need to change to read_csv with csv file or run a Pip install of Openpyxl
BRCA2_UG_AN = pd.read_excel("C:/Users/knigh/Documents/UHD MSDA/Capstone/Concordance/2026_BRCA2_ugCompleted_S3_.xlsx")
BRCA2_UG_AN.head()
#this file is pretty big dimension wise. The file I am using has 72 columns. It might we helpful to run this script in Spyder or VScode to see the data in full if needed

#Data Preview removed to comply with All of Us data user policies. You can see all the data's feature in the code block below.
```

```python
#since the file has 72 columns, lets only save the columns we are interested in
#listing the column names
BRCA2_UG_AN.columns
```

```python
#creating a subset of only the desired columns
BRCA2_SS = BRCA2_UG_AN[["vid","consequence","clinvar_classification","variant_type","dbsnp_rsid","revel", '(HumDiv) Polyphen Significance(S1)', 'PolyP_Score(S1)',
       'Sensitivity(S1)', 'FPR Score(S1)', 'Specificity(S1)',
       '(HumDiv) Polyphen Significance(S2)', 'PolyP_Score(S2)',
       'Sensitivity(S2)', 'FPR Score(S2)', 'Specificity(S2)',
       '(HumDiv) Polyphen Significance(S3)', 'PolyP_Score(S3)',
       'Sensitivity(S3)', 'FPR Score(S3)', 'Specificity(S3)']]

#brought down to 21 columns
BRCA2_SS.head() 

#There are no AOU participant identitifers in the preview below
```

```python
#checking for NAs
BRCA2_SS.isna().sum()

#we have 65 missing clinvar classifications, 34 rsids, and many missing student entered values
#we will handle these downstream when cleaning the data
```

## Aligning the columns values for comparisons

```python
#checking the unique values in the consequence column. Should all be variations of missense
BRCA2_SS.consequence.unique()
```

```python
#just changing all the missense to be missense_variant
BRCA2_SS.loc[BRCA2_SS["consequence"].str.contains("missense"), "consequence"] = "missense_variant"

#checking it worked
print(BRCA2_SS.consequence.unique())

#grabbing the count of missense values, should be the same as the # of rows
(BRCA2_SS['consequence']=='missense_variant').sum() #906
```

```python
#looking at the values in the clinvar classification columns
print(BRCA2_SS.clinvar_classification.unique())

#looking at the count of the values so when we condense them, we can make sure the count was allocated correctly
BRCA2_SS["clinvar_classification"].value_counts()
```

```python
#going to condense the classifcation labels into 4 groups - VUS, Disease Causing, Benign, and NA
#most labels do not have a mix of benign and pathogenic, so we can group them better. If you do have a mix, I would look at the predictions and see were they best align

#breakdown of grouping
#VUS - anything with uncertain in the label (will run this first)
#Disease Causing - pathogenic| likely pathogenic, pathogenic|benign, likely pathogenic, pathogenic| benign, likely benign, likely pathogenic
#Benign - |benign, likely benign | benign | likely benign |
#NA - Not Provided (run last) 

#creating a duplicate of Clinvar_classification to make these changes to, so we can keep the original column
BRCA2_SS["clinvar_CL"] = BRCA2_SS["clinvar_classification"]
BRCA2_SS[["clinvar_CL","clinvar_classification"]].head(5)
```

```python
#changing the labels of the clinvar_CL (concordance labels) to match the groups made eariler
#starting with the uncertain group
BRCA2_SS.loc[BRCA2_SS["clinvar_CL"].str.contains("uncertain", na=False), "clinvar_CL"] = "VUS"

#checking the counts, should be 776
BRCA2_SS["clinvar_CL"].value_counts()
```

```python
#now changing the pathogenic rows to be disease causing
BRCA2_SS.loc[BRCA2_SS["clinvar_CL"].str.contains("pathogenic", na=False), "clinvar_CL"] = "Disease Causing"

#checking count, there should be 5
BRCA2_SS["clinvar_CL"].value_counts()
```

```python
#now changing the benign group
BRCA2_SS.loc[BRCA2_SS["clinvar_CL"].str.contains("benign",na=False),"clinvar_CL"] = 'Benign'

#checking count, should be 60
BRCA2_SS["clinvar_CL"].value_counts()
```

```python
#Lastly changing the Not Provide to an NA
BRCA2_SS.loc[BRCA2_SS["clinvar_CL"].str.contains("not provided",na=False),"clinvar_CL"] = pd.NA

#counting NAs, should be 65
print(BRCA2_SS["clinvar_CL"].isna().sum())

#checking full count
BRCA2_SS["clinvar_CL"].value_counts()

#total is 906, same # of rows
```

```python
#Now updating the REVEL column
#REVEL has a score scale for their classifications:
#Revel score break down.
#≥ 0.644	Pathogenic supporting
#0.643 - 0.291	Neutral (it is benign leaning, so for concordance purposes we will label as benign)
#≤ 0.290	Benign supporting

#Copying the Revel Column to modify
BRCA2_SS['revel_CL'] = BRCA2_SS['revel']
BRCA2_SS[['revel_CL',"revel"]].head(5)
```

```python
#Updating the revel_CL column with the concordance labels

#assigning the revel scores to their labels based on the score scale
BRCA2_SS.loc[BRCA2_SS['revel'] >= .644, 'revel_CL'] = 'Disease Causing'
BRCA2_SS.loc[BRCA2_SS['revel'] <= .643, 'revel_CL'] = 'Benign'

print("Count of Disease causing is", (BRCA2_SS['revel_CL'] == 'Disease Causing').sum())
print("Count of Benign is",(BRCA2_SS['revel_CL'] == 'Benign').sum())

#60 Disease causing, 846 Benign
```

```python
#Moving on to the Polyphen columns to update them to CLs
#Poly-phen has 3 catagories - Benign, possibly damaging, and probably damanging. We will spilt into two groups Benign and Disease Causing

#copying the three Poly-phen classification columns to modify starting with S1 (student 1)
BRCA2_SS["(HumDiv) Polyphen Significance(S1)_CL"] = BRCA2_SS["(HumDiv) Polyphen Significance(S1)"]
BRCA2_SS[["(HumDiv) Polyphen Significance(S1)_CL", "(HumDiv) Polyphen Significance(S1)"]].head(5)
```

```python
#now S2 (student 2)
BRCA2_SS["(HumDiv) Polyphen Significance(S2)_CL"] = BRCA2_SS["(HumDiv) Polyphen Significance(S2)"]
BRCA2_SS[["(HumDiv) Polyphen Significance(S2)_CL", "(HumDiv) Polyphen Significance(S2)"]].head(5)
```

```python
#now S3 (S3)
BRCA2_SS["(HumDiv) Polyphen Significance(S3)_CL"] = BRCA2_SS["(HumDiv) Polyphen Significance(S3)"]
BRCA2_SS[["(HumDiv) Polyphen Significance(S3)_CL", "(HumDiv) Polyphen Significance(S3)"]].head(5)
```

```python
#updating the Poly-phen classifications to CLs

#starting with (HumDiv) Polyphen Significance(S1)
#We already have a benign classification, so we only need to updating the damaging classes to disease causing
BRCA2_SS.loc[BRCA2_SS["(HumDiv) Polyphen Significance(S1)"].str.lower().str.contains("Damaging",na=False),"(HumDiv) Polyphen Significance(S1)_CL"] = 'Disease Causing'
#not sure hy Damaging won't capture them all, but using both upper and lower case seems to work
BRCA2_SS.loc[BRCA2_SS["(HumDiv) Polyphen Significance(S1)"].str.lower().str.contains("damaging",na=False),"(HumDiv) Polyphen Significance(S1)_CL"] = 'Disease Causing'
BRCA2_SS.loc[BRCA2_SS["(HumDiv) Polyphen Significance(S1)"].str.lower().str.contains("benign",na=False),"(HumDiv) Polyphen Significance(S1)_CL"] = 'Benign'

#Next (HumDiv) Polyphen Significance(S2)
BRCA2_SS.loc[BRCA2_SS["(HumDiv) Polyphen Significance(S2)"].str.lower().str.contains("Damaging",na=False),"(HumDiv) Polyphen Significance(S2)_CL"] = 'Disease Causing'
BRCA2_SS.loc[BRCA2_SS["(HumDiv) Polyphen Significance(S2)"].str.lower().str.contains("damaging",na=False),"(HumDiv) Polyphen Significance(S2)_CL"] = 'Disease Causing'
BRCA2_SS.loc[BRCA2_SS["(HumDiv) Polyphen Significance(S2)"].str.lower().str.contains("benign",na=False),"(HumDiv) Polyphen Significance(S2)_CL"] = 'Benign'

#Next (HumDiv) Polyphen Significance(S3)
BRCA2_SS.loc[BRCA2_SS["(HumDiv) Polyphen Significance(S3)"].str.lower().str.contains("Damaging",na=False),"(HumDiv) Polyphen Significance(S3)_CL"] = 'Disease Causing'
BRCA2_SS.loc[BRCA2_SS["(HumDiv) Polyphen Significance(S3)"].str.lower().str.contains("damaging",na=False),"(HumDiv) Polyphen Significance(S3)_CL"] = 'Disease Causing'
BRCA2_SS.loc[BRCA2_SS["(HumDiv) Polyphen Significance(S3)"].str.lower().str.contains("benign",na=False),"(HumDiv) Polyphen Significance(S3)_CL"] = 'Benign'


#comparing counts
BRCA2_SS["(HumDiv) Polyphen Significance(S1)_CL"].value_counts()
```

```python
BRCA2_SS["(HumDiv) Polyphen Significance(S2)_CL"].value_counts()
```

```python
BRCA2_SS["(HumDiv) Polyphen Significance(S3)_CL"].value_counts()
```

## Comparing Corcordance Labels between Poly-phen, in-silco tools, and Cinvar

## Comparing Poly-phen student scores first

```python
#taking student 1, 2 and 3 poly-phen columns that we updated eariler with concordance labels and comparing them against eachother

PP_CL_Total = (
    (BRCA2_SS["(HumDiv) Polyphen Significance(S1)_CL"] == BRCA2_SS["(HumDiv) Polyphen Significance(S3)_CL"]) &
    (
        BRCA2_SS["(HumDiv) Polyphen Significance(S2)_CL"].isna() | #this prevents the missing values in S2 to count as discordance since there was alot
        (BRCA2_SS["(HumDiv) Polyphen Significance(S2)_CL"] == BRCA2_SS["(HumDiv) Polyphen Significance(S1)_CL"])
    )
).sum()

print(PP_CL_Total)

#904 concordant predictions
```

```python
#looking at the discordance between students
cols = [
    "(HumDiv) Polyphen Significance(S1)_CL",
    "(HumDiv) Polyphen Significance(S2)_CL",
    "(HumDiv) Polyphen Significance(S3)_CL"
]

# Rows where the non-missing values do not all agree
rows_disagree = BRCA2_SS[BRCA2_SS[cols].nunique(axis=1) > 1]

rows_disagree[cols].head(5)

#looked at variants in rows 672 and 752 and looked them up on polyPhen-2 separately
#672: Disease Causing
#752: Benign

#very weird, I double checked the Polyphen full report and confirmed S3 values, but when I input the variants manually one by one, the value was different depending if input as a single submission or batch
```

### Comparing Polyphen to Clinvar

```python
#since the labels for S1 were more acurrate, I will just use the S1 column for comparision
print((BRCA2_SS["clinvar_CL"] == BRCA2_SS["(HumDiv) Polyphen Significance(S1)_CL"]).sum())

#looking at the discordance
BRCA2_SS.loc[(BRCA2_SS["clinvar_CL"].isin(["Benign", "Disease Causing"])) & 
    (BRCA2_SS["clinvar_CL"] != BRCA2_SS["(HumDiv) Polyphen Significance(S1)_CL"]),["clinvar_CL", "(HumDiv) Polyphen Significance(S1)_CL"]]

#Clinvar and Polyphen align for 48 values. That is 73.85% of clinvar that is not VUS or missing
```

```python
#looking at the concordance between Clinvar and Polyphen
ClinPolconcord = BRCA2_SS.loc[BRCA2_SS["clinvar_CL"] == BRCA2_SS["(HumDiv) Polyphen Significance(S1)_CL"],["clinvar_CL", "(HumDiv) Polyphen Significance(S1)_CL"]]
ClinPolconcord.value_counts()
```

### Comparing PolyPhen to Revel

```python
#Finding how many time Polyphen labels match Clinvar labels
print((BRCA2_SS["revel_CL"] == BRCA2_SS["(HumDiv) Polyphen Significance(S1)_CL"]).sum())

#looking at the discordance
BRCA2_SS.loc[(BRCA2_SS["revel_CL"] != BRCA2_SS["(HumDiv) Polyphen Significance(S1)_CL"]),
             ["revel_CL", "(HumDiv) Polyphen Significance(S1)_CL"]]

#596 concordant, 65.78% concordant, 310 discordant
```

```python
#looking at the concordance between polyphen and REVEL
Wmodelconcord = BRCA2_SS.loc[(BRCA2_SS["revel_CL"] == BRCA2_SS["(HumDiv) Polyphen Significance(S1)_CL"]),["revel_CL", "(HumDiv) Polyphen Significance(S1)_CL"]]
Wmodelconcord.value_counts()
```

### Comparing Revel to Clinvar

```python
#Finding how many time REVEL labels match Clinvar labels
print((BRCA2_SS["clinvar_CL"] == BRCA2_SS["revel_CL"]).sum())

#looking at the discordance
BRCA2_SS.loc[(BRCA2_SS["clinvar_CL"].isin(["Benign", "Disease Causing"])) & #this line is here to skip over the VUS classifications
    (BRCA2_SS["clinvar_CL"] != BRCA2_SS["revel_CL"]),["clinvar_CL", "revel_CL"]]

#clinvar and revel align for 63 non-VUS/missing values out of 65, 96.9% concordant!!
```

```python
#looking at the concordance between Clinvar and REVEL
ClinRevconcord = BRCA2_SS.loc[BRCA2_SS["clinvar_CL"] == BRCA2_SS["revel_CL"],["clinvar_CL", "revel_CL"]]
ClinRevconcord.value_counts()
```

### Adding the created labels back onto the original df for exporting to model with in another script
